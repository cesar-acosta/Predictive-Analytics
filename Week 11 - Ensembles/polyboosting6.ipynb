## Gradient Boosting on Polynomial data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
df = pd.read_csv('dataset.csv')
df[:4]
```

```python
y = df.y
X = df.drop(['y'],axis =1)
```

```python
# boundary for x and y axes
axes=[-0.5, 0.5, -0.1, 0.8]
plt.plot(X, y,'k.')
plt.axis(axes)
plt.grid();
```

```python
# define a sequence of x-coordinate values
x1 = np.linspace(-0.5,0.5, 500)
# convert it to a matrix
x1 = x1.reshape(-1, 1)
```

```python
# build the regression tree
```

```python
tree = DecisionTreeRegressor(max_depth=2, random_state=42)
```

### Model one

```python
tree.fit(X, y);
red_line1 = tree.predict(x1)
```

```python
plt.plot(X,y,"k.",markersize=4)
plt.plot(x1,red_line1,"r-", linewidth=2)
plt.axis(axes)
plt.grid();
```

```python
yhat1 = tree.predict(X)
```

### Model two

```python
e2 = y - yhat1
tree.fit(X, e2)
red_line2 = red_line1 + tree.predict(x1)
```

```python
plt.plot(X, y, "k.",markersize=4)
plt.plot(x1, red_line2,"r-", linewidth=2)
plt.axis(axes)
plt.grid();
```

```python
yhat2 = tree.predict(X)
```

### Model three

```python
e3 = e2 - yhat2
tree.fit(X, e3);
red_line3 = red_line2 + tree.predict(x1)
```

```python
plt.plot(X, y, "k.",markersize=4)
plt.plot(x1,red_line3,"r-", linewidth=2)
plt.axis(axes)
plt.grid();
```

```python
yhat3 = tree.predict(X)
```

### Model four

```python
e4 = e3 - yhat3
tree.fit(X,e4)
red_line4 = red_line3 + tree.predict(x1)
yhat4 = tree.predict(X)
```

```python
plt.plot(X, y, "k.",markersize=4)
plt.plot(x1,red_line4,"r-", linewidth=2)
plt.axis(axes)
plt.grid();
```

## Using GradientBoostingRegressor( ) with 4 steps

```python
gbrt = GradientBoostingRegressor(max_depth=2, 
                                 n_estimators=4, 
                                 learning_rate=1.0, 
                                 random_state=42)
gbrt.fit(X, y)
y_pred = gbrt.predict(x1)
plt.plot(X, y, "k.",markersize=4)
plt.plot(x1, y_pred,"r-",linewidth=2)
plt.axis(axes)
plt.grid();
```

### with 20 steps

```python
gbrt2 = GradientBoostingRegressor(max_depth=2, 
                                  n_estimators=20, 
                                  learning_rate=0.1, 
                                  random_state=42)
gbrt2.fit(X, y);
y_pred = gbrt2.predict(x1)
plt.plot(X,y,"k.",markersize=4)
plt.plot(x1, y_pred,"r-",linewidth=2)
plt.axis(axes)
plt.grid();
```

### with 50 steps

```python
gbrt2 = GradientBoostingRegressor(max_depth=2, 
                                  n_estimators=50, 
                                  learning_rate=0.1, 
                                  random_state=42)
gbrt2.fit(X, y);
y_pred = gbrt2.predict(x1)
plt.plot(X,y,"k.", label=None,markersize=4)
plt.plot(x1,y_pred,"r-",linewidth=2)
plt.axis(axes)
plt.grid();
```

### with 80 steps

```python
gbrt2 = GradientBoostingRegressor(max_depth=2, 
                                  n_estimators=80, 
                                  learning_rate=0.1, 
                                  random_state=42)
gbrt2.fit(X, y);
y_pred = gbrt2.predict(x1)
plt.plot(X,y,"k.",markersize=4)
plt.plot(x1, y_pred,"r-",linewidth=2)
plt.axis(axes)
plt.grid();
```
