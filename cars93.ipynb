## groupby and apply

```python
# Select five cars with largest Price values
```

```python
# function selecting five rows with largest numerical col  (McKinney Ch 10.3)
```

```python
def top5 (df,n=5,column = 'Price'):
    return df.sort_values(by=column)[-n:]
```

```python
# five most expensive cars by DriveTrain
```

```python
df.groupby('DriveTrain').apply(top5)
```

```python
# Manufacturer and Model of five most expensive cars by DriveTrain
```

```python
df4=df.groupby('DriveTrain').apply(top5)
```

```python
df5 = df4[['Manufacturer','Model','Price']]
```

```python
df5
```
